```python
import pandas as pd
import numpy as np
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_cities_by_average_temperature').text

from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'html.parser')
print(soup.prettify())
```

```python
table = soup.find('table', {'class':'wikitable'})
```

```python
import awoc

c = awoc.AWOC().get_countries_list()
c
```

```python
links= table.find_all('a')
links
```

```python
countries = []
cities = []
for link in links:
    if link.get('title') in c:
        countries.append(link.get('title'))
    elif link.get('title') != None:
        cities.append(link.get('title'))
countries,cities
```

```python
Africa = pd.DataFrame()
Africa['Countries'] = countries + ['x', 'x']
Africa['Cities'] = cities
print(Africa)
```
